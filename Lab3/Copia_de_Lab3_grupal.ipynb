```python
# read the sequence lines for be able to search in it
with open('sequence_Covid-19.fasta') as f:
    lines = f.readlines()
print(lines)
```

Usar dirección '/**/464/' del genoma corregido, pues la dirección altual es local.

```python
# join the lines in a single string for represent the sequence
secuence=''
for l in lines[1:]:
  secuence+=l
print(secuence)
```

Ejemplo Bioinformático de busqueda por metodo de fuerza bruta


```python
# initialize target sequence data and make a example for determine the actual sequence position
target = 'AATGAATTCTCGTAACTA'
a = secuence.find(target)
print(a)
```

```python
# travel trough the sequence positions one by one until the last possible position
# for compare the target with a slice of the sequence in that position until find it
# finally prints the sequence and his start and final position
for x in range(len(secuence) - len(target)):
  if secuence[x: x + len(target)] == target:
    print(secuence[x: x + len(target)])
    print(x)
    print(x + len(target))
    break
```
